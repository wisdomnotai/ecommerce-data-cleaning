```python
#impoting the pandas library

import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing and loading the dataset

sales_data = pd.read_csv("dataset/raw_sales_data_1000_rows.csv")
sales_data.info()
sales_data.head()
sales_data.describe()
```

```python
#checking for missing values

sales_data.isnull().sum()
```

```python
#filling missing data

sales_data["Phone"] = sales_data["Phone"].fillna("not available")
sales_data["Quantity"] = sales_data["Quantity"].fillna(1)
sales_data["Email"] = sales_data["Email"].fillna("unkownemail@gmail.com")
sales_data.isnull().sum()
```

```python
#checking for duplicates

sales_data.duplicated().any()
sales_data.duplicated().sum()
```

```python
#Checking for incorrect datatypes

sales_data.info()
sales_data.head(10)
```

```python
#deleting unnecessary columns

sales_data = sales_data.drop(["CustomerName", "Phone", "Email"], axis = 1)
sales_data.info()
```

```python
#fixing incorrect datatypes

sales_data["OrderID"] = pd.to_numeric(sales_data["OrderID"])
sales_data["State"] = sales_data["State"].str.upper()
sales_data["Price"] = sales_data["Price"].astype(str)
sales_data["Price"] = sales_data["Price"].str.replace(",", "", regex = True)
sales_data["Price"] = pd.to_numeric(sales_data["Price"])
sales_data["Product"] = sales_data["Product"].str.lower()
sales_data["Quantity"] = pd.to_numeric(sales_data["Quantity"])
sales_data["OrderDate"] = pd.to_datetime(sales_data["OrderDate"], errors = "coerce", dayfirst = True)
sales_data["PaymentStatus"] = sales_data["PaymentStatus"].str.lower()
sales_data.head(10)
```

```python
#saving the dataset to csv

sales_data.to_csv("dataset/sales_data_clean.csv", index = False)
```
