```python
#importing the pandas library

import pandas as pd
```

```python
#importing the dataset

sales_data = pd.read_csv("dataset/raw_sales_data_1000_rows.csv")
```

```python
#Getting data statistics
sales_data.describe()
```

```python
#checking for missing values

sales_data.isnull().sum()
```

```python
#filling missing data in the dataset

sales_data["Phone"] = sales_data["Phone"].fillna("not available")
sales_data["Quantity"] = sales_data["Quantity"].fillna(1)
sales_data["Email"] = sales_data["Email"].fillna("unkwoknemail@gmail.com")
```

```python
sales_data.isnull().sum()
```

```python
#fixing incorrect datatypes

sales_data["OrderID"] = pd.to_numeric(sales_data["OrderID"])
sales_data["Phone"] = sales_data["Phone"].astype(str).str.replace(r'\D', "", regex=True)
sales_data["Price"] = sales_data["Price"].replace({',': '', '.': ''}, regex = True)
sales_data["Quantity"] = pd.to_numeric(sales_data["Quantity"])
```

```python
#removing duplicates

sales_data = sales_data.drop_duplicates()
```

```python
#fixing messy texts

#removing extra spaces from the data column
sales_data["CustomerName"] = sales_data["CustomerName"].str.strip().str.replace(r'\s+', ' ', regex=True)

#making all texts upper case in the states column
sales_data["State"] = sales_data["State"].str.upper()

#making all texts in product lower
sales_data["Product"] = sales_data["Product"].str.lower()

#making all texts in PaymentStatus lower
sales_data["PaymentStatus"] = sales_data["PaymentStatus"].str.lower()

#making all texts in email lower
sales_data["Email"] = sales_data["Email"].str.lower()
```

```python
#saving the newly cleaned dataset
sales_data.to_csv("dataset/cleaned_sales_data.csv",index = False)
```
